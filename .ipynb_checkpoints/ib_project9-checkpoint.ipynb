```python
import psycopg2 as pg
import getpass
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
user = input("Username: ")
password = getpass.getpass("Password: ")
conn = pg.connect(dbname='csci403', user=user, password=password, host='codd.mines.edu', port='5433')
del user
del password
cur = conn.cursor()
```

```python
query = """SELECT tsla.day, tsla.price tsla_price, btc.price btc_price FROM tsla, btc WHERE tsla.day = btc.day;"""
cur.execute(query)
sql_table = cur.fetchall()
df = pd.DataFrame(sql_table, columns=['day', 'tsla_price', 'btc_price'])
df
```

```python
np.corrcoef(df['tsla_price'], df['btc_price'])
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(df['day'], df['btc_price'], color='blue')
ax[1].plot(df['day'], df['tsla_price'], color='red')
```

```python
query = """SELECT * FROM btc;"""
cur.execute(query)
sql_table = cur.fetchall()
df = pd.DataFrame(sql_table, columns=['day', 'price', 'total_btc', 'hash_rate', 'difficulty', 'output_volume'])
df
```

```python
fig, ax = plt.subplots(5)
for i, name in enumerate(df.columns[1:]):
    ax[i].plot(df['day'], df[name])
    ax[i].set_title(name, y=0.5)
```
